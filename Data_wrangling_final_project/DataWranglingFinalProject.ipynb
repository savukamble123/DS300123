```python
import pandas as pd
import numpy as np
```

```python
# 1. Import the dataset using Pandas from the given URL
url = "https://raw.githubusercontent.com/SR1608/Datasets/main/covid-data.csv"
df = pd.read_csv(url)
```

```python
# 2. High Level Data Understanding
# a. Find the number of rows & columns in the dataset
num_rows, num_cols = df.shape
print("Number of rows:", num_rows)
print("Number of columns:", num_cols)
```

```python
# b. Data types of columns
print("Data types of columns:")
print(df.dtypes)
```

```python
# c. Info & describe of data in the dataframe
print("Info:")
print(df.info())
print("Describe:")
print(df.describe())
```

```python
# 3. Low Level Data Understanding
# a. Count of unique values in the 'location' column
unique_locations = df['location'].nunique()
print("Number of unique locations:", unique_locations)
```

```python
# b. Continent with the maximum frequency using value counts
continent_frequency = df['continent'].value_counts()
max_continent = continent_frequency.idxmax()
print("Continent with maximum frequency:", max_continent)
```

```python
# c. Maximum and mean value in 'total_cases'
max_total_cases = df['total_cases'].max()
mean_total_cases = df['total_cases'].mean()
print("Maximum total cases:", max_total_cases)
print("Mean total cases:", mean_total_cases)
```

```python
# d. Quartile values (25%, 50%, 75%) in 'total_deaths'
quartiles = df['total_deaths'].quantile([0.25, 0.5, 0.75])
print("Quartile values for total deaths:")
print(quartiles)
```

```python
# e. Continent with the maximum 'human_development_index'
max_hdi_continent = df.loc[df['human_development_index'].idxmax(), 'continent']
print("Continent with maximum human development index:", max_hdi_continent)
```

```python
# f. Continent with the minimum 'gdp_per_capita'
min_gdp_continent = df.loc[df['gdp_per_capita'].idxmin(), 'continent']
print("Continent with minimum GDP per capita:", min_gdp_continent)
```

```python
# 4. Filter the dataframe with specific columns and update the dataframe
selected_columns = ['continent', 'location', 'date', 'total_cases', 'total_deaths', 'gdp_per_capita', 'human_development_index']
df = df[selected_columns]
df
```

```python
# 5. Data Cleaning
# a. Remove duplicate observations
df.drop_duplicates(inplace=True)
df
```

```python
# b. Find missing values in all columns
missing_values = df.isnull().sum()
print("Missing values in each column:")
print(missing_values)
```

```python
# c. Remove observations where the continent column value is missing
df.dropna(subset=['continent'], inplace=True)
```

```python
# d. Fill missing values with 0
df.fillna(0, inplace=True)
```

```python
# 6. Date time format
# a. Convert the date column to datetime format
df['date'] = pd.to_datetime(df['date'])
```

```python
# b. Create a new column 'month' after extracting month data from the date column
df['month'] = df['date'].dt.month
```

```python
# 7. Data Aggregation
# a. Find the max value in all columns using groupby function on 'continent'
df_groupby = df.groupby('continent').max().reset_index()
```

```python
# 8. Feature Engineering
# a. Create a new feature 'total_deaths_to_total_cases' by the ratio of 'total_deaths' to 'total_cases'
df_groupby['total_deaths_to_total_cases'] = df_groupby['total_deaths'] / df_groupby['total_cases']
```

```python
# 9. Data Visualization
# a. Univariate analysis on 'gdp_per_capita' column using seaborn dist plot
import seaborn as sns
import matplotlib.pyplot as plt
sns.histplot(df_groupby['gdp_per_capita'])
plt.show()
```

```python
# b. Scatter plot of 'total_cases' & 'gdp_per_capita'
sns.scatterplot(data=df_groupby, x='total_cases', y='gdp_per_capita')
plt.show()
```

```python
# c. Pairplot on df_groupby dataset
sns.pairplot(df_groupby)
plt.show()
```

```python
# d. Bar plot of 'continent' column with 'total_cases'
sns.catplot(data=df_groupby, x='continent', y='total_cases', kind='bar')
plt.show()
```
